```python
import pandas as pd
```

```python
dataframe = pd.read_csv("android_bids_us.csv")
```

```python
print(dataframe.isnull().sum())
```

```python
dataframe.dropna(inplace = True)
```

```python
c_bidid = 'bidid'
c_utc_time = 'utc_time'
c_app_id = 'app_id'
c_user_state = 'user_state'
c_user_isp = 'user_isp'
c_device_maker = 'device_maker'
c_device_model = 'device_model'
c_device_osv = 'device_osv'
c_device_height = 'device_height'
c_device_width = 'device_width'
c_marketplace = 'marketplace'
c_click = 'click'
```

```python
train_features = [c_utc_time, c_app_id, c_user_state, c_user_isp, c_device_maker,
                  c_device_model, c_device_osv, c_device_height, c_device_width]
target_feature = c_click
```

```python
print(dataframe.info())
```

```python
from sklearn.preprocessing import LabelEncoder

categorical_attrs = dataframe.select_dtypes(include='object').columns.difference([c_bidid, c_app_id])

for label in categorical_attrs:
    encoder = LabelEncoder()
    dataframe[label] = encoder.fit_transform(dataframe[label])
```

```python
dataframe
```

```python
from sklearn.model_selection import train_test_split

test_data_fraction = 0.33

validation_fraction_of_train = 0.20

def split_train_test_validation(df):
    df_train, df_test = train_test_split(df, test_size = test_data_fraction,random_state=1)
    df_aux = df_train.copy()
    df_train, df_validation = train_test_split(df_aux, test_size = validation_fraction_of_train,random_state=1)
    return df_train, df_test.copy(), df_validation
```

```python
df_train, df_test, df_validation = split_train_test_validation(dataframe)

X_train = df_train[train_features]
y_train = df_train[target_feature]

X_test = df_test[train_features]
y_test = df_test[target_feature]

X_validation = df_validation[train_features]
y_validation = df_validation[target_feature]
```
